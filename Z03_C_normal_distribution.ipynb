# Wielowymiarowy rozkład normalny

```python
import matplotlib.pylab as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.mlab as mlab
from matplotlib import cm
```

# Zad. 
Narysujmy gęstość dwuwymiarowego rozkładu Normalnego o parametach 

* n=[0,0], cov = [[1,0],[0,1]]
* n=[0,0], cov = [[2,0],[0,1]]
* n=[0,0], cov = [[1,0],[0,2]]

```python
from scipy.stats import multivariate_normal

n=[0,0]
cov=[[1,0],[0,1]]
x_axis = np.linspace(-5,5,500)
y_axis = np.linspace(-5,5,500)
X, Y = np.meshgrid(x_axis,y_axis)
x = multivariate_normal(n, cov)
cov=[[2,0],[0,1]]
x1 = multivariate_normal(n, cov)
cov=[[1,0],[0,2]]
x2 = multivariate_normal(n, cov)
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X; pos[:, :, 1] = Y
fig = plt.figure(figsize=(5,5))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, x.pdf(pos))
ax.plot_surface(X, Y, x1.pdf(pos))
ax.plot_surface(X, Y, x2.pdf(pos))


plt.show()
```

# Zad 
Narysujmy kontury gęstości dwuwymiarowego rozkładu Normalnego o parametach 

* n=[0,0], cov = [[1,0],[0,1]]
* n=[0,0], cov = [[2,0],[0,1]]
* n=[0,0], cov = [[1,0],[0,2]]

```python
n = [0, 0]
cov = [[1, 0], [0, 1]]
fig, ax = plt.subplots(nrows=1,ncols=1, figsize=(5,5))
x_axis = np.linspace(-3,3,500)
y_axis = np.linspace(-3,3,500)
X, Y = np.meshgrid(x_axis,y_axis)
x = multivariate_normal(n, cov)
cov = [[2, 0], [0, 1]]
x1 = multivariate_normal(n, cov)
cov = [[1, 0], [0, 2]]
x2 = multivariate_normal(n, cov)
pos = np.dstack((X,Y))
ax.contour(X, Y, x.pdf(pos))
ax.contour(X, Y, x1.pdf(pos))
ax.contour(X, Y, x2.pdf(pos))
plt.show()
```
